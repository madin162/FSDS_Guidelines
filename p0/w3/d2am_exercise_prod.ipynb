```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd

%matplotlib inline
```

## Create PDFs

```python
def add_wnoise(signal):
    return signal + np.random.normal(0,scale=abs(0.2*np.maximum(signal,0.05)),size=len(signal))
```

```python
x = np.arange(0,5,0.01)
```

```python
from scipy.stats import expon, norm, uniform, gamma
fig, ax = plt.subplots(1, 1)
functions = [expon.pdf,norm.pdf,uniform.pdf, gamma.pdf]
pdfs=[]
keywords=[{},{'loc':2},{'loc':1,'scale':1.5},{'a':2}]
colors=['r','g','b','m','y']
for idx,fn in enumerate(functions):
    pdf = add_wnoise(np.apply_along_axis(fn,0,x,**keywords[idx]))
    pdf /= (0.01 * pdf).sum()
    pdfs.append(pdf)
    

for idx in range(len(functions)):
    ax.plot(x, pdfs[idx],colors[idx]+'-', lw=1, alpha=0.6, label='expon pdf')
    #ax.plot(x, norm_pdf,'b-', lw=1, alpha=0.6, label='expon pdf')
    
pdfs = np.array(pdfs).T
```

```python
dfs = np.hstack([np.expand_dims(x,-1),pdfs])
dfs = pd.DataFrame(dfs,columns=['x','p1','p2','p3','p4'])
```

```python
dfs.to_csv('data\data_pdfs.csv',index=False)
```
